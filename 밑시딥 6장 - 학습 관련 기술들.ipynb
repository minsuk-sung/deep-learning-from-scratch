### 모험가 이야기

![](https://image.slidesharecdn.com/dloptimization-170715221703/95/optimization-in-deep-learning-1-638.jpg?cb=1500157166)

> 색다른 모험가가 있습니다. 광할한 메마른 산맥을 여행하면서 날마다 깊은 골짜기를 찾아 발걸음을 옮깁니다. 그는 전설에 나오는 세상에서 가장 깊고 낮은 골짜기, `깊은 곳`을 찾아가려 합니다. 그것이 그의 여행의 목적이죠. 게다가 그는 `엄격한 제약` 2개로 자신을 옭아맸습니다.

> 하나는 지도를 보지 않는 것  
> 또 하나는 눈가리개를 쓰는 것  

> 지도도 없고 보이지도 않으니 가장 낮은 골짜기가 광대한 땅 어디에 있는지 도저히 알 도리가 없죠. 그런 혹독한 조건에서 이 모험가는 어떻게 `깊은 곳`을 찾을 수 있을까요? 어떻게 걸음을 옮겨야 효율적으로 찾아낼 수 있을까요?

- 이러한 어려운 상황에서도 가장 중요한 단서 : `땅의 기울기`
- 아무리 주위 경치는 볼 수 없지만 지금 서 있는 땅의 기울기는 알 수 있음
- 이걸 통해서 가장 크게 기울어진 방향으로 가자는 것이 SGD의 전략

### 확률적 경사 하강법(SGD)

$$ W \leftarrow W - \eta \frac{\partial L}{\partial W} $$


```python
# 최적화를 담당하는 클래스를 분리해 구현하면 기능을 모듈화하기 좋음
class SGD:

    def __init__(self, lr=0.01):
        self.lr = lr # 문제 : Q. 왜 학습률을 저장하는 인스턴스를 만들어뒀을까? A. Learing Rate Decay
    
    # 실제로 이렇게 함수로 만들어두면 나중에 편하게 최적화 클래스를 바꿀 수 있음
    def update(self, params, grads):
        for key in params.keys():
            params[key] -= self.lr * grads[key] 
            
network = TwoLayerNet(...)
optimizer = SGD() # 나중에 이걸 SGD() 가 아니라 Momentum() 이렇게도 가능하단 소리

for i in range(10000):
    ...
    x_batch, t_batch = get_mini_batch(...)
    grads = network.gradient(x_batch, t_batch)
    params = network.params
    optimizer.update(params, grads) # 이렇게 할 수 있다는 소리임
```

### 확률적 경사 하강법(SGD)의 단점
- 단순하고 구현도 쉽지만 단점이 있음
- 단점 : 비효율적

### Momentum

- 모멘텀 : `momentum`, 물리학에서 `운동량`을 나타내는 단어

![](https://postfiles.pstatic.net/MjAxNzA3MjdfMjY3/MDAxNTAxMTIzMjgxNzM3.w9b6TOLwpgUiewZxUZURx6fqnkNIeEK74p6JJ7JzTxwg.yiFTwTm_cPvGVkcBJlT13igEPmNAoFXAXMeTrLWcE2cg.PNG.cjswo9207/fig_6-4.png?type=w1)

$$ v \leftarrow \alpha v - \eta \frac{\partial L}{\partial W} $$

$$ W \leftarrow W + v $$  

<br>
<center>( $v$: 속도, $W$: 갱신할 가중치 매개변수, $\frac{\partial L}{\partial W}$: $W$에 대한 손실 함수의 기울기, $\eta$: 학습률 )</center>

- $\alpha v$ : 물체가 아무런 힘을 받지 않을 때 `서서히 하강시키는` 역할, 마치 지면 마찰이나 공기저항과 같은 역할이라고 볼 수 있음, $\alpha$는 0.9 등의 값으로 설정함
- 첫번째 식은 기울기 방향으로 힘을 받아 물체가 가속된다는 물리법칙을 수식화한 것

````python
class Momentum:

    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None # 물체의 속도, 초기화 때는 아무 값도 담지 않음
        
    def update(self, params, grads):
        if self.v is None: # 매개변수와 같은 구조의 데이터를 딕셔너리 변수로 저장함
            self.v = {}
            for key, val in params.items():                                
                self.v[key] = np.zeros_like(val)
                
        for key in params.keys():
            # 첫번째 식
            self.v[key] = self.momentum*self.v[key] - self.lr*grads[key] 
            # 두번째 식
            params[key] += self.v[key] 
````

- 모멘텀의 갱신 경로는 공이 그릇 바닥을 구르듯 움직임
- x축의 힘은 아주 작지만 방향이 변하지 않아서 한 방향으로 일정하게 가속됨
- y축의 힘은 크지만, 위아래로 번갈아 받아서 상충하여 y축 방향의 속도는 안정적이지 않음
- 지그재그의 정도가 SGD보다 덜한 것을 알 수 있음, SGD보다 x축 방향으로 빠르게 다가가 지그재그 움직임이 줄어듬

### AdaGrad

- 학습률(Learning Rate)는 신경망 학습에서의 성공을 결정하는 중요한 하이퍼 파라미터
- 학습률이 너무 작으면 학습시간이 너무 길어지고
- 학습률이 너무 크면 발산하여 제대로 학습이 이루어지지 않음

> 학습률을 고정하지말고, 점차 줄여나가면서 학습을 할 수 있을까?  
⇒ 학습률 감소(Learning Rate Decay) : 처음에는 크게 학습하다가 조금씩 작게 학습해보자

- 학습률 감소 기법을 적용하기 위한 가장 간단한 방법은 **매개변수 `전체`의 학습률 값을 일괄적으로 낮춰보는** 방법!!
- 이러한 아이디어를 발전시킨 방법이 AdaGrad

> AdaGrad : `각각의` 매개변수에 `맞춤형` 값을 만들어줌

- 개별 매개변수에 적응적으로(`ada`ptive) 학습률을 조정하면서 학습을 진행(`grad`ient)

$$ h \leftarrow h + \frac{\partial L}{\partial W} \odot \frac{\partial L}{\partial W}$$

$$ W \leftarrow W - \eta \frac{1}{\sqrt{h}}\frac{\partial L}{\partial W} $$

<br>
<center>( $W$: 갱신할 가중치 매개변수, $\frac{\partial L}{\partial W}$: 은 $W$에 대한 손실함수의 기울기, $\eta$: 학습률, $h$: 기존 기울기 값을 제곱하여 계속 더해줌, $\odot$: 행렬의 원소별 곱셈 )</center>

- 매개변수를 갱신할 때, $\frac{1}{\sqrt{h}}$ 를 곱해 학습률을 조정
- 매개변수의 원소 중에서 많이 움직인(크게 갱신된) 원소는 학습률이 낮아지고, 적게 움직인(적게 움직인) 원소는 학습률이 크게 갱신됨
- 다시 말해서, 학습률 감소가 매개변수의 원소마다 다르게 적용됨

````python
class AdaGrad:

    def __init__(self, lr=0.01):
        self.lr = lr
        self.h = None
        
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
            
        for key in params.keys():
            # 첫번째 식
            self.h[key] += grads[key] * grads[key]
            # 두번째 식
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7) # 0으로 나누는 사태를 방지하기 위해, 작은 값 더해줌
````

- y축 방향으로는 처음에는 크게 움직이지만, 그 큰 움직임에 비례해 갱신정도도 큰 폭으로 작아지도록 조정됨
- 그래서 y축 방향으로 갱신 강도가 빠르게 약해지고, 지그재그 움직임이 줄어듬

### Adam

- 모멘텀 : 공기 그릇 바닥을 구르는 듯한 움직임
- AdaGrad : 매개변수의 원소마다 적응적으로 갱신 정도를 조정
- Adam : **모멘텀와 AdaGrad를 적절하게 융합**해보면 어떨까?
    - 매개변수 공간을 효율적으로 탐색
    - 하이퍼 파라미터의 편향보정이 진행
    
    
$$ m_t \leftarrow m_t + \frac{m_t}{1-\beta_1^t} $$

$$ v_t \leftarrow v_t + \frac{v_t}{1-\beta_2^t} $$

$$ W \leftarrow W - \frac{\eta}{\sqrt{v_t}+\epsilon} m_t $$

````python
class Adam:

    def __init__(self, lr=0.001, beta1=0.9, beta2=0.999):
        self.lr = lr       # Adam에서 쓰이는 하이퍼 파라미터 : 학습률
        self.beta1 = beta1 # Adam에서 쓰이는 하이퍼 파라미터 : 일차 모멘텀용 계수, default: 0.9
        self.beta2 = beta2 # Adam에서 쓰이는 하이퍼 파라미터 : 이차 모멘텀용 계수, default: 0.999
        self.iter = 0
        self.m = None
        self.v = None
        
    def update(self, params, grads):
        if self.m is None:
            self.m, self.v = {}, {}
            for key, val in params.items():
                self.m[key] = np.zeros_like(val)
                self.v[key] = np.zeros_like(val)
        
        self.iter += 1
        lr_t  = self.lr * np.sqrt(1.0 - self.beta2**self.iter) / (1.0 - self.beta1**self.iter)         
        
        for key in params.keys():
            #self.m[key] = self.beta1*self.m[key] + (1-self.beta1)*grads[key]
            #self.v[key] = self.beta2*self.v[key] + (1-self.beta2)*(grads[key]**2)
            self.m[key] += (1 - self.beta1) * (grads[key] - self.m[key])
            self.v[key] += (1 - self.beta2) * (grads[key]**2 - self.v[key])
            
            params[key] -= lr_t * self.m[key] / (np.sqrt(self.v[key]) + 1e-7)
            
            #unbias_m += (1 - self.beta1) * (grads[key] - self.m[key]) # correct bias
            #unbisa_b += (1 - self.beta2) * (grads[key]*grads[key] - self.v[key]) # correct bias
            #params[key] += self.lr * unbias_m / (np.sqrt(unbisa_b) + 1e-7)
````

- 모멘텀과 비슷한 패턴, 모멘텀 때보다 공의 좌우 흔들림이 적음
- 이는 학습의 갱신 강도를 적응적으로 조정해서 얻은 혜택

### Nesterov Momentum → 책에 없는 내용

참고 : http://newsight.tistory.com/224

- 모멘텀에서 한 단계 더 발전된 방법으로 다음과 같은 수식을 통해 W 업데이트가 이루어진다.

$$ V \leftarrow \mu V - \lambda f'(W+\mu V) $$

$$ W \leftarrow W - V $$

- 현재의 W에서의 gradient를 계산하기 전에, 먼저 학습했던 V만큼 한 번 더 가본 상태(W+µV)에서 gradient를 계산하는 것
- 이것은 예를 들자면, 관성에 좀 더 힘을 주어 일단 먼저 갔던 속도만큼 가본다음에, gradient를 계산해서 방향을 조정하겠다는 것이
- 결국 보다 공격적으로 **속도벡터 V를 활용하겠다**는 것
- 나머지는 모멘텀처럼 그전 V를 일정부분 반영해서 업데이트를 한다
- 아래의 그림과 논문을 보면 좀더 직관적인 이해가 가능하다.
![](https://t1.daumcdn.net/cfile/tistory/2302D049562BD6C023)

````python
class Nesterov:
    
    def __init__(self, lr=0.01, momentum=0.9):
        self.lr = lr
        self.momentum = momentum
        self.v = None
        
    def update(self, params, grads):
        if self.v is None:
            self.v = {}
            for key, val in params.items():
                self.v[key] = np.zeros_like(val)
            
        for key in params.keys():
            self.v[key] *= self.momentum
            self.v[key] -= self.lr * grads[key]
            params[key] += self.momentum * self.momentum * self.v[key]
            params[key] -= (1 + self.momentum) * self.lr * grads[key]
````

### RMSProp  → 책에서는 자세히 언급X

- AdaGrad는 과거의 기울기를 제곱하여 계속 더해나감
- 이러한 방법은 학습이 진행할수록 갱신 강도가 약해지므로 이를 계속 진행하면 어느 순간 갱신량이 0이 되어 전혀 갱신되지 않게 됨
- 이를 개선한 방법이 RMSProp
    - RMSProp는 **먼 과거의 기울기는 서서히 잊고 새로운 기울기 정보를 크게 반영**하자
    - 이를 지수이동평균(Exponential Moving Average, EMA)라 하며, 과거 기울기의 반영 규모를 기하급수적으로 감소시킴
    

$$ E[g^2]_t \leftarrow 0.9E[g^2]_t + 0.1g_t^2 $$

$$ \theta \leftarrow \theta - \frac{\eta}{\sqrt{E[g^2]_t + \epsilon}} g_t$$

````python
class RMSprop:

    def __init__(self, lr=0.01, decay_rate = 0.99):
        self.lr = lr
        self.decay_rate = decay_rate
        self.h = None
        
    def update(self, params, grads):
        if self.h is None:
            self.h = {}
            for key, val in params.items():
                self.h[key] = np.zeros_like(val)
            
        for key in params.keys():
            self.h[key] *= self.decay_rate
            self.h[key] += (1 - self.decay_rate) * grads[key] * grads[key]
            params[key] -= self.lr * grads[key] / (np.sqrt(self.h[key]) + 1e-7)
````

## 가중치의 초깃값 (p202)

### 초깃값을 0으로 하면?

- 가중치 감소(weight decay)
    - 오버피팅을 억제해 범용 성능을 높이는 테크닉
    - 가중치 매개변수의 값이 작아지도록 학습하는 방법

- 가중치를 작게 만들고 싶으면 초깃값도 최대한 작은 값에서 시작하는 것이 일단 정공법

> 애초에 그냥 **`초깃값을 모두 0으로 설정`** 해버리면 되지 않을까?
- 결론부터 말하자면 최악의 아이디어, 학습이 올바르게 이뤄지지 않을 것
- 가중치를 균일한 값으로 설정하면 오차역전파법에서 **`모든 가중치의 값이 똑같이 갱신`** 되기 때문에
- 이렇게 되면 가중치를 여러 개 갖는 의미가 사라지게 됨

- 가중치가 고르게 되어버리는 상황을 막으려면(가중치의 대칭적인 구조를 무너뜨리려면) `초깃값을 무작위로 설정`해야함

### 은닉층의 활성화값 분포

- 은닉충의 활성화값의 분포를 관찰하면 중요한 정보를 얻을 수 있음

> **Xavier 초깃값** : 앞 계층의 노드가 n개 일 때, 초깃값의 표준편차가 <font color='red'> $\frac{1}{\sqrt{n}}$ </font> 이 되도록 설정, 단 n은 앞 계층의 노드 수

- 현재 다수의 딥러닝 프레임워크들이 표준적으로 이용함
- 각 층의 활성화값들은 광범위하게 분포시킬 목적으로 가중치의 적절한 분포를 찾고자 함
- 그러다가 앞 계층의 노드가 n개이면 표준편차가 $\frac{1}{\sqrt{n}}$인 분포를 사용하면 된다는 결론을 이끌었음

![](https://postfiles.pstatic.net/MjAxNzA3MjdfMjc0/MDAxNTAxMTI2ODI0ODM3.LTKJgqWpZsgRgHX2-IhP5EM6MUo8tOTk_Fzl89bVV48g.ER6261bifFkOBBgxEwAamC3eDWBTABYd1r4gpXxrICsg.PNG.cjswo9207/fig_6-12.png?type=w1)

> **He 초깃값** : 앞 계층의 노드가 n개 일 때, 초깃값의 표준편차가 <font color='red'> $\sqrt{\frac{2}{n}}$ </font> 이 되도록 설정

- Xavier 초깃값 방법에 비해서, ReLU를 사용하게 될 경우에는 `음의 영역이 0이 되어버림`
- 그렇기 때문에 초깃값을 더 넓게 분포시키기 위해서 `2배의 계수가 필요하다고 직감적`으로 해석할 수 있음

````python
def show_hidden_distribution(w, act_func):    
    input_data = np.random.randn(1000, 100)  # 1000개의 데이터
    node_num = 100  # 각 은닉층의 노드(뉴런) 수
    hidden_layer_size = 5  # 은닉층이 5개
    activations = {}  # 이곳에 활성화 결과를 저장

    x = input_data

    for i in range(hidden_layer_size):
        if i != 0:
            x = activations[i-1]
        a = np.dot(x, w)
        z = act_func(a)
        activations[i] = z

    # 히스토그램 그리기
    for i, a in activations.items():
        plt.subplot(1, len(activations), i+1)
        plt.title(str(i+1) + "-layer")
        if i != 0: plt.yticks([], [])
        plt.hist(a.flatten(), 30, range=(0,1))
    plt.show()
````

````python
# 초깃값을 다양하게 바꿔가며 실험해보자！
init_w = np.random.randn(node_num, node_num)
w_list = {'표준편차가 1인 정규분포로 초기화': init_w * 1,
          '표준편차가 0.01인 정규분포로 초기화': init_w * 0.01,
          '[ Xavier 초깃값 ] 표준편차가 1/sqrt(n)인 정규분포로 초기화': init_w * np.sqrt(1.0 / node_num),
          '[ He 초깃값 ] 표준편차가 sqrt(2/n)인 정규분포로 초기화': init_w * np.sqrt(2.0 / node_num)}
````

#### 활성화함수가 sigmoid 함수일 경우

````python
# 첫번째 활성화 함수 : 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
````

````python
for method,w in w_list.items():
    print('{}를 이용했을 경우'.format(method))
    show_hidden_distribution(w, sigmoid)
````

- 표준편차가 1인 정규분포로 초기화를 이용했을 경우
    - 각 층의 활성화값들이 0과 1에 치우쳐 분포
    - 시그모이드 함수는 출력이 0에 가까워지자(혹은 1에 가까워지자) 미분이 0에 다가감
    - 즉, 0과 1에 치우쳐 분포하게 되면 역전파의 기울기 값이 점점 작아지다가 사라짐 ⇒ 기울기 소실 (Vanishing Gradient) 문제 발생
    - 참고로 그 반대로 기울기가 발산하는 경우는 기울기 발산 (Exploding Gradient)라고 함, 이것 또한 문제가 되긴 함
- 표준편차가 0.01인 정규분포로 초기화를 이용했을 경우
    - 0.5 부근에 집중됨
    - 활성화 값들이 치우쳤다는 것은 표현력의 관점에서는 큰 문제
    - 즉, 다수의 뉴런이 거의 같은 값을 출력하고 있으니 뉴런을 여러개 둔 의미가 없어진다는 뜻
    - 활성화값들이 치우치면 표현력을 제한한다는 관점에서 문제가 됨

> **`각 층의 활성화값은 적당히 고루 분포`** 되어야함  
> 층과 층 사이에 적당하게 다양한 데이터가 흐르게 해야 신경망 학습이 효율적으로 이뤄지기 때문  
> 반대로 치우친 데이터가 흐르면 기울기 소실이나 표현력 제한 문제가 생겨서 학습이 잘 이뤄지지 않음

- [ Xavier 초깃값 ] 표준편차가 1/sqrt(n)인 정규분포로 초기화를 이용했을 경우
    - 앞 층에 노드가 많을수록 대상 노드의 초깃값으로 설정하는 가중치가 좁게 퍼짐
    - 층이 깊어지면서 형태가 다소 일그러지지만, 앞의 2가지 방식보다는 확실히 넓게 분포됨을 확인 가능
    - 그러므로 각 층에 흐르는 데이터는 적당히 퍼저 있으므로, 시그모이드 함수의 표현력도 제한 받지 않고, 학습이 효율적으로 이뤄질 것
    > 다소 일그러지는 문제는 사실 시그모이드 함수가 애초에 원점 대칭 함수가 아니기 때문  
    그러므로 활성화 함수용으로는 원점에서 대칭인 함수가 바람직하다고 알려져 있음

#### 활성화함수가 ReLU 함수일 경우

````python
# 두번째 활성화 함수 : ReLU 함수
def ReLU(x):
    return np.maximum(0, x)
````

````python
for method,w in w_list.items():
    print('{}를 이용했을 경우'.format(method))
    show_hidden_distribution(w, ReLU)
````

- [ He 초깃값 ] 표준편차가 sqrt(2/n)인 정규분포로 초기화를 이용했을 경우
    - 이번에는 활성화 함수로 ReLU를 사용
    - Xavier 초깃값은 활성화 함수가 선형인 것을 전제로 이끈 결과, 따라서 sigmoid와 tanh 함수와 같이 좌우 대칭인 즉, 중앙 부근이 선형인 함수에 적절하다고 볼 수 있음
    - 그러나, ReLU를 이용할 경우에는 ReLU에 특화되어 있는 초깃값을 사용해야함
    - ReLU를 활성화 함수로 사용할 경우, He 초깃값 사용해야함
    - 신경망에 아주 작은 데이터가 흐른다는 것은 역전파 때 가중치의 기울기 역시 아주 작아진다늗 뜻인데, `He 초깃값 방법을 제외한 나머지 방법은 층이 깊어질수록 활성화값들의 치우침이 점점 커지고, 학습할 때 기울기 소실 문제가 발생`됨
    - 반면, He 초깃값은 그 분포가 치우치지 않고 유지됨을 볼 수 있음

#### 활성화함수가 tanh 함수일 경우

````python
# 세번째 활성화 함수 : tanh 함수
def tanh(x):
    return np.tanh(x)
````

````python
for method,w in w_list.items():
    print('{}를 이용했을 경우'.format(method))
    show_hidden_distribution(w, tanh)
````

## 배치 정규화 (p210)

> 각 층이 활성화를 적당히 퍼뜨리도록 강제해보면 어떨까?  
이러한 아이디어에서 출발한 방법이 `배치 정규화(Batch Normalization)`

> 잠깐! 헷갈리지 말자!  
<b><font color='red' size = 8>$$Normalization \neq Regularization$$</font></b>

### 배치 정규화 알고리즘

- 각 층에서의 활성화값이 적당히 분포되도록 조정하는 것
- 아래와 같이 데이터 분포를 정규화하는 배치 정규화 계층을 신경망에 삽입함

![](https://postfiles.pstatic.net/MjAxNzA3MjdfOTUg/MDAxNTAxMTI4NTM3NTU2.aLAMgxDvmG9clAXUJ47gnN4DWmxvSr93uk5p_n6gDyMg.DqUt_OKyr-zNjpk_I7q6421W5aL7-NEhj8aTCMkWvXgg.PNG.cjswo9207/fig_6-16.png?type=w1)

- 학습시 미니배치를 단위로 정규화
- 데이터 분포가 평균이 0이고 분산이 1이 되도록 정규화함

<center><font size = 5>평균$\mu_B$</font></center>

$$ \mu_B \leftarrow \frac{1}{m}\sum_{i=1}^m x_i $$
<br>

<center><font size = 5>분산$\sigma_B^2$</font></center>

$$ \sigma_B^2 \leftarrow \frac{1}{m}\sum_{i=1}^m (x_i - \mu_B)^2 $$
<br>

<center><font size = 5>정규화하는 부분</font></center>
$$ \hat x_i \leftarrow \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}} $$

- 미니배치 $B=\{ x_1, x_2, ... , x_m \}$이라는 $m$개의 입력 데이터의 집합에 대해서 평균 $\mu_B$와 분산 $\sigma_B^2$을 구함
- 그리고 입력 데이터를 평균이 0이고 분산 1이 되게 (즉, 적절한 분포가 되게) 정규화
- 이 처리는 활성화 함수의 앞 혹은 뒤에 삽입함으로써 데이터 분포가 덜 치우치게 할 수 있음

- 배치 정규화 계층마다 이 정규화된 데이터에 고유한 확대(scale)와 이동(shift) 변환을 수행

$$ y_i \leftarrow \gamma \hat x_i + \beta $$

<center>($\gamma$가 확대, $\beta$가 이동)</center>

- 배치 정규화 계층을 계산그래프로 표현하면 다음과 같음

![](https://postfiles.pstatic.net/MjAxNzA3MjdfMjYx/MDAxNTAxMTI5MDU1Mzg0.lsuyZE7qxn42sz10-EPOhwx0pTaXF9X0o8UCjI4LDS8g.nl-o2tc5upQvairyHQUU1RjZg5aUz2e9xu5KfbIHjBwg.PNG.cjswo9207/fig_6-17.png?type=w1)

````python
class BatchNormalization:

    def __init__(self, gamma, beta, momentum=0.9, running_mean=None, running_var=None):
        self.gamma = gamma
        self.beta = beta
        self.momentum = momentum
        self.input_shape = None # 합성곱 계층은 4차원, 완전연결 계층은 2차원  

        # 시험할 때 사용할 평균과 분산
        self.running_mean = running_mean
        self.running_var = running_var  
        
        # backward 시에 사용할 중간 데이터
        self.batch_size = None
        self.xc = None
        self.std = None
        self.dgamma = None
        self.dbeta = None

    def forward(self, x, train_flg=True):
        self.input_shape = x.shape
        if x.ndim != 2:
            N, C, H, W = x.shape
            x = x.reshape(N, -1)

        out = self.__forward(x, train_flg)
        
        return out.reshape(*self.input_shape)
            
    def __forward(self, x, train_flg):
        if self.running_mean is None:
            N, D = x.shape
            self.running_mean = np.zeros(D)
            self.running_var = np.zeros(D)
                        
        if train_flg:
            mu = x.mean(axis=0)
            xc = x - mu
            var = np.mean(xc**2, axis=0)
            std = np.sqrt(var + 10e-7)
            xn = xc / std
            
            self.batch_size = x.shape[0]
            self.xc = xc
            self.xn = xn
            self.std = std
            self.running_mean = self.momentum * self.running_mean + (1-self.momentum) * mu
            self.running_var = self.momentum * self.running_var + (1-self.momentum) * var            
        else:
            xc = x - self.running_mean
            xn = xc / ((np.sqrt(self.running_var + 10e-7)))
            
        out = self.gamma * xn + self.beta 
        return out

    def backward(self, dout):
        if dout.ndim != 2:
            N, C, H, W = dout.shape
            dout = dout.reshape(N, -1)

        dx = self.__backward(dout)

        dx = dx.reshape(*self.input_shape)
        return dx

    def __backward(self, dout):
        dbeta = dout.sum(axis=0)
        dgamma = np.sum(self.xn * dout, axis=0)
        dxn = self.gamma * dout
        dxc = dxn / self.std
        dstd = -np.sum((dxn * self.xc) / (self.std * self.std), axis=0)
        dvar = 0.5 * dstd / self.std
        dxc += (2.0 / self.batch_size) * self.xc * dvar
        dmu = np.sum(dxc, axis=0)
        dx = dxc - dmu / self.batch_size
        
        self.dgamma = dgamma
        self.dbeta = dbeta
        
        return dx
````

### 배치 정규화의 효과

> 학습을 빨리 진행할 수 있음 (학습 속도 개선)  
> 초깃값에 크게 의존하지 않음 (골치 아픈 초깃값 선택 개선)  
> 오비피팅을 억제함 (드롭아웃 등의 필요성 감소)

![](https://jovianlin.io/content/images/2018/12/scaling.png)

````python
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

# 학습 데이터를 줄임
x_train = x_train[:1000]
t_train = t_train[:1000]

max_epochs = 20
train_size = x_train.shape[0]
batch_size = 100
learning_rate = 0.01
````

````python
def __train(weight_init_std):
    bn_network = MultiLayerNetExtend(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,
                                     weight_init_std=weight_init_std, use_batchnorm=True)
    network = MultiLayerNetExtend(input_size=784, hidden_size_list=[100, 100, 100, 100, 100], output_size=10,
                                  weight_init_std=weight_init_std)
    optimizer = SGD(lr=learning_rate)

    train_acc_list = []
    bn_train_acc_list = []

    iter_per_epoch = max(train_size / batch_size, 1)
    epoch_cnt = 0

    for i in range(1000000000):
        batch_mask = np.random.choice(train_size, batch_size)
        x_batch = x_train[batch_mask]
        t_batch = t_train[batch_mask]

        for _network in (bn_network, network):
            grads = _network.gradient(x_batch, t_batch)
            optimizer.update(_network.params, grads)

        if i % iter_per_epoch == 0:
            train_acc = network.accuracy(x_train, t_train)
            bn_train_acc = bn_network.accuracy(x_train, t_train)
            train_acc_list.append(train_acc)
            bn_train_acc_list.append(bn_train_acc)

            #print("epoch:" + str(epoch_cnt) + " | " + str(train_acc) + " - " + str(bn_train_acc))

            epoch_cnt += 1
            if epoch_cnt >= max_epochs:
                break

    return train_acc_list, bn_train_acc_list
````

````python
# 그래프 그리기==========
weight_scale_list = np.logspace(0, -4, num=16)
x = np.arange(max_epochs)
plt.figure(figsize=(16,16))

for i, w in enumerate(weight_scale_list):
    #print( "============== " + str(i+1) + "/16" + " ==============")
    train_acc_list, bn_train_acc_list = __train(w)
    
    plt.subplot(4,4,i+1)
    plt.title("W:" + str(w))
    if i == 15:
        plt.plot(x, bn_train_acc_list, label='Batch Normalization', markevery=2)
        plt.plot(x, train_acc_list, linestyle = "--", label='Normal(without BatchNorm)', markevery=2)
    else:
        plt.plot(x, bn_train_acc_list, markevery=2)
        plt.plot(x, train_acc_list, linestyle="--", markevery=2)

    plt.ylim(0, 1.0)
    if i % 4:
        plt.yticks([])
    else:
        plt.ylabel("accuracy")
    if i < 12:
        plt.xticks([])
    else:
        plt.xlabel("epochs")
plt.legend(loc='lower right')
plt.show()
````

> 거의 모든 경우에서 배치 정규화를 사용할 때의 학습 진도가 빠른 것이 나타남  
배치 정규화를 이용하지 않는 경우엔 초깃값이 잘 분포되어 있지 않으면 학습이 전혀 진행되지 않는 모습도 확인가능

## 바른 학습을 위해 (p215)

- 기계학습에서는 오버피팅과의 싸움
> 오버피팅(Overfitting) : 신경망이 훈련 데이터에만 지나치게 적응되어 그 외의 데이터에는 제대로 대응하지 못하는 상태

````python
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

# 오버피팅을 재현하기 위해 학습 데이터 수를 줄임
x_train = x_train[:300]
t_train = t_train[:300]
````

### 오버피팅

- 오버피팅은 주로 두가지 경우에 대해서 일어남
    - (1) 매개변수가 많고 표현력이 높은 모델을 쓸 경우
    - (2) 훈련 데이터가 적을 경우
    
- MNIST 데이터 중 훈련 데이터를 300개만(1) 사용하고 7층 네트워크(2)를 사용해서 예시를 들어보자

````python
network = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100, 100, 100], output_size=10)
optimizer = SGD(lr=0.01) # 학습률이 0.01인 SGD로 매개변수 갱신

max_epochs = 201
train_size = x_train.shape[0]
batch_size = 100

train_loss_list = []
train_acc_list = []
test_acc_list = []

iter_per_epoch = max(train_size / batch_size, 1)
epoch_cnt = 0

for i in range(1000000000):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]

    grads = network.gradient(x_batch, t_batch)
    optimizer.update(network.params, grads)

    if i % iter_per_epoch == 0:
        train_acc = network.accuracy(x_train, t_train)
        test_acc = network.accuracy(x_test, t_test)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)

        # print("epoch:" + str(epoch_cnt) + ", train acc:" + str(train_acc) + ", test acc:" + str(test_acc))

        epoch_cnt += 1
        if epoch_cnt >= max_epochs:
            break


# 그래프 그리기==========
markers = {'train': 'o', 'test': 's'}
x = np.arange(max_epochs)
plt.plot(x, train_acc_list, marker='o', label='train', markevery=10)
plt.plot(x, test_acc_list, marker='s', label='test', markevery=10)
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.ylim(0, 1.0)
plt.legend(loc='lower right')
plt.show()
````

- 훈련 데이터를 사용해서 측정한 정확도는 100 epoch을 지나는 무렵부터는 거의 100%
- 하지만 시험 데이터에서는 큰 차이를 보임
- 즉, 오버피팅이 되었다는 소리임

### 가중치 감소(weight decay)

- 오버피팅의 억제용으로 가장 많이 이용해온 방법 중 하나가 **`가중치 감소(weight decay)`**
- 학습 과정에서 큰 가중치에 대해서는 그에 상응하는 큰 불이익을 줘서 오버피팅을 억제하는 방법
- 보통 오버피팅이 일어나는 이유는 가중치 매개변수으이 값이 커서 발생하는 경우가 많기 때문

- 신경망 학습의 목적은 손실함수의 값을 최소화하는 것
- 그럼 이 때, 손실함수의 값에 **`가중치의 제곱 노름(L2 norm)`**를 더함, 이렇게 하면 가중치가 커지는 것을 억제할 수 있음
- 즉, 가중치를 $W$라 하면, L2노름에 따른 가중치 감소는 $\frac{1}{2}\lambda W^2$가 되고 이 값을 손실함수에 더함
- 이 때 $\lambda$를 억제의 세기를 조절하는 하이퍼파라미터, 이 값이 커질수록 큰 가중치에 대한 페널티도 커질 것
- 사실 <font color='red'>Regularization</font>를 다르게 말한 것이었음

![](https://t1.daumcdn.net/cfile/tistory/99B7603359820B9228)

````python
# weight decay（가중치 감쇠） 설정 =======================
#weight_decay_lambda = 0 # weight decay를 사용하지 않을 경우
weight_decay_lambda = 0.1
# ====================================================

network = MultiLayerNet(input_size=784, hidden_size_list=[100, 100, 100, 100, 100, 100], output_size=10,
                        weight_decay_lambda=weight_decay_lambda)
optimizer = SGD(lr=0.01) # 학습률이 0.01인 SGD로 매개변수 갱신

max_epochs = 201
train_size = x_train.shape[0]
batch_size = 100

train_loss_list = []
train_acc_list = []
test_acc_list = []

iter_per_epoch = max(train_size / batch_size, 1)
epoch_cnt = 0

for i in range(1000000000):
    batch_mask = np.random.choice(train_size, batch_size)
    x_batch = x_train[batch_mask]
    t_batch = t_train[batch_mask]

    grads = network.gradient(x_batch, t_batch)
    optimizer.update(network.params, grads)

    if i % iter_per_epoch == 0:
        train_acc = network.accuracy(x_train, t_train)
        test_acc = network.accuracy(x_test, t_test)
        train_acc_list.append(train_acc)
        test_acc_list.append(test_acc)

        #print("epoch:" + str(epoch_cnt) + ", train acc:" + str(train_acc) + ", test acc:" + str(test_acc))

        epoch_cnt += 1
        if epoch_cnt >= max_epochs:
            break


# 그래프 그리기==========
markers = {'train': 'o', 'test': 's'}
x = np.arange(max_epochs)
plt.plot(x, train_acc_list, marker='o', label='train', markevery=10)
plt.plot(x, test_acc_list, marker='s', label='test', markevery=10)
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.ylim(0, 1.0)
plt.legend(loc='lower right')
plt.show()
````

- Regularization 을 적용한 결과
- 훈련 데이터와 시험 데이터의 정확도의 차이가 줄었음
- 오버피팅이 억제

### 드롭아웃

- 신경망이 복잡해질수록 가중치 감소만으로는 대응하기 힘들어짐
- 그래서 나온 기법이 드롭아웃(Dropout)

> 드롭아웃(Dropout) : 뉴런을 임의로 삭제하면서 학습하는 방법

- 훈련 때 은닉층의 뉴런을 무작위로 골라 삭제
- 삭제된 뉴런은 아래 그림과 같이 신호를 전달하지 않게 됨
- 훈련 때는 데이터를 흘릴 때마다 삭제할 뉴런을 무작위로 선택하고
- 시험 때는 모든 뉴런에 신호를 전달, 단 시험때는 각 뉴런의 출력에 훈련 때 삭제한 비율을 곱하여 출력

![](https://postfiles.pstatic.net/MjAxNzA3MjdfMjEz/MDAxNTAxMTMzMzE4MDAz.ZAl6mcZ7VvWhLR82U6dovtoNeLbbmFHCGupK5hcJICAg.Z0wpOnQUwMLHuddn3XdFhgT-ghnzT8NHzZMbB4pFWcIg.PNG.cjswo9207/fig_6-22.png?type=w1)

````python
class Dropout:

    def __init__(self, dropout_ratio=0.5):
        self.dropout_ratio = dropout_ratio
        self.mask = None

    
    def forward(self, x, train_flg=True): # 여기를 통해서 드롭아웃을 설정
        if train_flg:
            # 핵심은 훈련 시에는 순전파 때마다 self.mask에 삭제할 뉴런을 False로 표시
            # self.mask 는 x와 형상이 같은 배열을 무작위로 생성하고, 그 값이 dropout_ratio 보다 큰 원소만 True로 설정
            self.mask = np.random.rand(*x.shape) > self.dropout_ratio
            return x * self.mask
        else:
            return x * (1.0 - self.dropout_ratio)

    def backward(self, dout):
        # 역전파시에는 ReLU랑 동작원리가 동일함
        # 순전파 때 신호를 통과시키는 뉴런은 역전파 때도 신호를 그대로 통과시킴
        # 순전파 때 통과시키지 않은 뉴런은 역전파 때도 신호를 차단
        return dout * self.mask
````

우리가 조정해줘야하는 여러가지 하이퍼파라미터가 존재하는데 이러한 값을 어떻게 최대한 효율적으로 탐색할까?

![](http://research.sualab.com/assets/images/bayesian-optimization-overview-1/unluck-in-manual-search-process.gif)

### 검증 데이터

- 지금까지는 데이터셋을 훈련 데이터와 시험 데이터 크게 2가지로 분리했음
- 주의해야할 점은 하이퍼파라미터의 성능을 평가하기 위해서 시험 데이터를 건들면 안됨 ( 오버피팅되기 때문 )
- 그런데 하이퍼파라미터를 다양한 값으로 설정하고 검증하기 위해서 데이터셋을 다르게 분리할 것

> **검증 데이터(validation data)** : 하이퍼 파라미터를 조정하기 위한 데이터, 보통 <font color='red'>훈련 데이터 중 20% 정도를 검증 데이터로</font> 먼저 분리해둠

- 결국 데이터셋은 이제 3가지 분류로 나뉘어짐
    1. 훈련 데이터 (train data) : <font color='red'>매개변수(가중치와 편향)</font>의 학습에 이용
    2. 검증 데이터 (validation data) : <font color='red'>하이퍼 파라미터</font>의 성능을 평가하는데 이용
    3. 시험 데이터 (test data) : <font color='red'>신경망의 범용 성능을 확인</font>하기 위해서 이용

### 하이퍼파라미터 최적화

- 하이퍼 파라미터를 최적화할 때의 핵심은 하이퍼 파라미터의 최적값이 존재하는 범위를 조금씩 줄여나가는 것
- **`신경망의 하이퍼 파라미터 최적화`**에서는 그리드 서치(grid search)같이 규칙적인 탐색보다는 **`무작위로 샘플링해 탐색하는 편이 더 좋다`**

```
[ 0단계 ]
    - 하이퍼 파라미터 값의 범위를 설정
    
[ 1단계 ]
    - 설정된 범위에서 하이퍼 파라미터의 값을 무작위로 추출
    
[ 2단계 ]
    - 1단계에서 샘플링한 하이퍼파라미터 값을 사용하여 학습
    - 검증 데이터로 정확도를 평가 (단, 에폭 작게)
    
[ 3단계 ]
    - 1,2단계를 특정 횟수 반복
    - 그에 대한 정확도 결과를 보고 하이퍼 파라미터의 범위를 좁힘
```

> 하지만 사실상 최적의 방법이 아닌 실용적인 방법일 뿐  
즉, 과학이라기보다는 다분히 앞선 수행자들의 지혜와 직관에 의존하는 느낌이 듬, 세련된 기법을 원한다면 베이즈 최적화(Bayesian Optimization)을 살펴보자

### 하이퍼파라미터 최적화 구현하기

````python
(x_train, t_train), (x_test, t_test) = load_mnist(normalize=True)

# 결과를 빠르게 얻기 위해 훈련 데이터를 줄임
x_train = x_train[:500]
t_train = t_train[:500]

# 20%를 검증 데이터로 분할
validation_rate = 0.20
validation_num = int(x_train.shape[0] * validation_rate)
x_train, t_train = shuffle_dataset(x_train, t_train) # 훈련 데이터를 분리하기 전에 뒤섞자
x_val = x_train[:validation_num]
t_val = t_train[:validation_num] # 검증 데이터
x_train = x_train[validation_num:]
t_train = t_train[validation_num:] # 검증 데이터의 라벨
````

````python
def __train(lr, weight_decay, epocs=50):
    
    network = MultiLayerNet(input_size=784,
                            hidden_size_list=[100, 100, 100, 100, 100, 100],
                            output_size=10,
                            weight_decay_lambda=weight_decay)
    
    trainer = Trainer(network, x_train, t_train, x_val, t_val,
                      epochs=epocs,
                      mini_batch_size=100,
                      optimizer='sgd',
                      optimizer_param={'lr': lr}, 
                      verbose=False)
    trainer.train()

    return trainer.test_acc_list, trainer.train_acc_list
````

````python
# 하이퍼파라미터 무작위 탐색======================================
optimization_trial = 100
results_val = {}
results_train = {}
for _ in range(optimization_trial):
    
    # 탐색한 하이퍼파라미터의 범위 지정===============
    weight_decay = 10 ** np.random.uniform(-8, -4)
    lr = 10 ** np.random.uniform(-6, -2)
    # ================================================

    val_acc_list, train_acc_list = __train(lr, weight_decay)
    # print("val acc:" + str(val_acc_list[-1]) + " | lr:" + str(lr) + ", weight decay:" + str(weight_decay))
    key = "lr:" + str(lr) + ", weight decay:" + str(weight_decay)
    results_val[key] = val_acc_list
    results_train[key] = train_acc_list
````

````python
# 그래프 그리기========================================================
print("=========== Hyper-Parameter Optimization Result ===========")
graph_draw_num = 20
col_num = 5
row_num = int(np.ceil(graph_draw_num / col_num))
i = 0

plt.figure(figsize=(10,10))

for key, val_acc_list in sorted(results_val.items(), key=lambda x:x[1][-1], reverse=True):
    print("Best-" + str(i+1) + "(val acc:" + str(val_acc_list[-1]) + ") | " + key)

    plt.subplot(row_num, col_num, i+1)
    plt.title("Best-" + str(i+1))
    plt.ylim(0.0, 1.0)
    if i % 5: plt.yticks([])
    plt.xticks([])
    x = np.arange(len(val_acc_list))
    plt.plot(x, val_acc_list)
    plt.plot(x, results_train[key], "--")
    i += 1

    if i >= graph_draw_num:
        break

plt.show()
````

* 가중치 초깃값을 정하는 방법은 올바른 학습을 하는 데 매우 중요하다.
* 가중치의 초깃값으로는 ‘Xavier 초깃값’과 ‘He 초깃값’이 효과적이다.
* 배치 정규화를 이용하면 학습을 빠르게 진행할 수 있으며, 초깃값에 영향을 덜 받게 된다.
* 오버피팅을 억제하는 정규화 기술로는 가중치 감소와 드롭아웃이 있다.
* 하이퍼파라미터 값 탐색은 최적 값이 존재할 법한 범위를 점차 좁히면서 하는 것이 효과적이다.
